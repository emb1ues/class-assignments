```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
def decisionBoundaryPlotterKNN(k, data, labels):
    
    #kNN
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(data,labels)
    
    #make grid
    X = np.linspace(-4,4,40)
    Y = np.linspace(-4,4,40)
    grid = np.transpose([np.tile(X, len(Y)), np.repeat(Y, len(X))])
    plt.scatter(grid[:,0],grid[:,1])
    
    #find labels for the grid points
    labels_ = [neigh.predict([[grid[i,0],grid[i,1]]]) for i in range(len(X)**2)]
    colors_ = list(map(lambda x: 'r' if(x == -1) else 'b', labels_))
    
    #plot
    plt.scatter(grid[:,0],grid[:,1],c = colors_)
```

```python
#training data
data = np.array([[-1,1],[0,1],[0,2],[1,-1],[1,0],[1,2],[2,2],[2,3]])
labels = [-1,+1,-1,-1,+1,+1,-1,+1]
colors = list(map(lambda x: 'r' if(x == -1) else 'b', labels))
```

```python
plt.scatter(data[:,0],data[:,1],c = colors)
```

```python
decisionBoundaryPlotterKNN(3, data, labels)
```

```python
decisionBoundaryPlotterKNN(5, data, labels)
```

```python
decisionBoundaryPlotterKNN(7, data, labels)
```
