```python
import numpy as np
from matplotlib import pyplot as plt
import sklearn.datasets
from sklearn.cluster import KMeans,DBSCAN
```

```python
def clusterPlotter(dataset,labels,centers):
    listColors = ['.b','.g','.r','.c','.m','.y','.k']
    
    #make a dictionary later
    listColors = listColors[0:centers]
    clusterID = np.unique(labels)
    
    for i in range(len(dataset)):
        x = dataset[i,0]
        y = dataset[i,1]
        
        if labels[i] == -1:
            c = '.k'
        else:
            for j in range(len(clusterID)):
                if labels[i] == clusterID[j]:
                    c = listColors[j]
        
        plt.plot(x,y,c)
    plt.show()
    
```

## Random Dataset

```python
dataset,real_labels = sklearn.datasets.make_blobs(100,2, centers=3)
plt.plot(dataset[:,0],dataset[:,1],'.k')
```

## KMeans

```python
centers = 3
kmeans = KMeans(n_clusters=centers).fit(dataset)
clusterPlotter(dataset, kmeans.labels_,centers)
```

## DBSCAN

```python
epsilon = 1.5
n = 5
dbscan = DBSCAN(eps=epsilon, min_samples=n).fit(dataset)
clusterPlotter(dataset,dbscan.labels_,len(np.unique(dbscan.labels_)))
```
